```python
# W8A1
class Cylinder:
    def __init__(self, r, h):
        self.r = r
        self.h = h
        self.pi = 3.14
    def S_sur(self):
        return f"{self.h * 2 * self.r * self.pi + 2 * self.pi * self.r:.2f}"
    def V(self):
        return f"{self.r ** 2 * self.pi * self.h:.2f}"
r, h = map(float, input().split())
ans = Cylinder(r, h)
print(ans.S_sur(), ans.V())
```

```python
# W8A2
class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year
    def namnhuan(self):
        if (self.year % 4 == 0 and self.year % 100 != 0) or (self.year % 400 == 0):
            return True
        return False
    def maxday(self):
        max_day = 0
        if self.month in [1, 3, 5, 7, 8, 10, 12]:
            max_day = 31
        elif self.month in [4, 6, 9, 11]:
            max_day = 30
        else:
            max_day = 29 if self.namnhuan() else 28
        return max_day
    def check(self):
        if self.month > 12:
            return False
        if self.day <= self.maxday():
            return True
        return False
    def next_day(self):
        if self.check() == False:
            return "INVALID"
        
        if self.day < self.maxday():
            return f"{self.day + 1}/{self.month}/{self.year}"
        elif self.month != 12:
            return f"{1}/{self.month + 1}/{self.year}"
        else:
            return f"{1}/{1}/{self.year + 1}"
d, m, y = map(int, input().split("/"))
cal = Date(d, m, y)
print(cal.next_day())
            
```

```python
# W8A3
import math

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def is_origin(self):
        return self.x == 0 and self.y == 0

    def is_on_y_axis(self):
        return self.x == 0

    def is_on_x_axis(self):
        return self.y == 0

    def distance_to(self, other):
        return math.sqrt((self.x - other.x)**2 + (self.y - other.y)**2)

x_input = float(input("Nhập hoành độ x của điểm M: "))
y_input = float(input("Nhập tung độ y của điểm M: "))

m = Point(x_input, y_input)
origin = Point(0, 0)

if m.is_origin():
    print("M là gốc tọa độ.")
else:
    print("M không phải là gốc tọa độ.")

if m.is_on_y_axis():
    print("M nằm trên trục tung.")
else:
    print("M không nằm trên trục tung.")

if m.is_on_x_axis():
    print("M nằm trên trục hoành.")
else:
    print("M không nằm trên trục hoành.")

dist = m.distance_to(origin)
print(f"Khoảng cách từ M tới gốc tọa độ là: {dist:.2f}")
```

```python
# W8A4
class Calculator:
    def __init__(self, n1, n2):
        self.num1 = n1
        self.num2 = n2

    def set_numbers(self, n1, n2):
        self.num1 = n1
        self.num2 = n2

    def add(self):
        return self.num1 + self.num2

    def subtract(self):
        return self.num1 - self.num2

    def product(self):
        return self.num1 * self.num2

    def divide(self):
        if self.num2 == 0:
            return "Lỗi: Không thể chia cho 0"
        return self.num1 / self.num2

    def power(self):
        return self.num1 ** self.num2

    def mod(self):
        if self.num2 == 0:
            return "Lỗi: Không thể chia lấy dư cho 0"
        return self.num1 % self.num2

n1 = float(input("Nhập số thứ nhất: "))
n2 = float(input("Nhập số thứ hai: "))
calc = Calculator(n1, n2)

while True:
    print("\n--- MENU MÁY TÍNH ---")
    print("1. Cộng (+)")
    print("2. Trừ (-)")
    print("3. Nhân (*)")
    print("4. Chia (/)")
    print("5. Lũy thừa (^)")
    print("6. Chia lấy dư (%)")
    print("7. Thay đổi 2 số mới")
    
    choice = input("Chọn phép tính (1-7): ")

    if choice == '1':
        print(f"Kết quả: {calc.add()}")
    elif choice == '2':
        print(f"Kết quả: {calc.subtract()}")
    elif choice == '3':
        print(f"Kết quả: {calc.product()}")
    elif choice == '4':
        print(f"Kết quả: {calc.divide()}")
    elif choice == '5':
        print(f"Kết quả: {calc.power()}")
    elif choice == '6':
        print(f"Kết quả: {calc.mod()}")
    elif choice == '7':
        n1 = float(input("Nhập số thứ nhất mới: "))
        n2 = float(input("Nhập số thứ hai mới: "))
        calc.set_numbers(n1, n2)
        print("Đã cập nhật số mới thành công!")
    else:
        print("Lựa chọn không hợp lệ.")

    quit_ask = input("\nBạn có muốn thoát? (Gõ 'yes', 'YES' hoặc 'Yes' để dừng): ")
    if quit_ask in ["yes", "YES", "Yes"]:
        print("Tạm biệt!")
        break
```

```python
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, name, price):
        product = {"name": name, "price": price}
        self.items.append(product)
        print(f"Đã thêm '{name}' vào giỏ hàng.")

    def remove_item(self, name):
        found = False
        for item in self.items:
            if item['name'].lower() == name.lower():
                self.items.remove(item)
                found = True
                print(f"Đã xóa sản phẩm '{name}' khỏi giỏ hàng.")
                break
        if not found:
            print(f"Không tìm thấy sản phẩm có tên '{name}'.")

    def is_empty(self):
        return len(self.items) == 0

    def calculate_total(self):
        total = sum(item['price'] for item in self.items)
        return total

    def display_cart(self):
        if self.is_empty():
            print("Giỏ hàng đang rỗng.")
        else:
            print("--- Danh sách sản phẩm trong giỏ ---")
            for i, item in enumerate(self.items, 1):
                print(f"{i}. {item['name']}: {item['price']:,} VNĐ")

    def clear_cart(self):
        self.items.clear()
        print("Đã xóa toàn bộ giỏ hàng.")

cart = ShoppingCart()

while True:
    print("2. Thêm sản phẩm mới")
    print("3. Xóa sản phẩm theo tên")
    print("4. Kiểm tra giỏ hàng rỗng")
    print("5. Tính tổng tiền")
    print("6. Hiển thị toàn bộ sản phẩm")
    print("7. Xóa toàn bộ giỏ hàng")
    
    choice = input("Nhập lựa chọn của bạn (2-7): ")

    if choice == '2':
        name = input("Nhập tên sản phẩm: ")
        try:
            price = float(input("Nhập giá sản phẩm: "))
            cart.add_item(name, price)
        except ValueError:
            print("Giá sản phẩm phải là một con số.")
    
    elif choice == '3':
        name = input("Nhập tên sản phẩm cần xóa: ")
        cart.remove_item(name)
        
    elif choice == '4':
        if cart.is_empty():
            print("Giỏ hàng đang rỗng.")
        else:
            print("Giỏ hàng đang có sản phẩm.")
            
    elif choice == '5':
        print(f"Tổng tiền hiện tại: {cart.calculate_total():,} VNĐ")
        
    elif choice == '6':
        cart.display_cart()
        
    elif choice == '7':
        cart.clear_cart()
    
    else:
        print("Lựa chọn không hợp lệ, vui lòng chọn từ 2 đến 7.")

    quit_ask = input("\nBạn có muốn thoát? (Gõ 'yes', 'YES' hoặc 'Yes' để dừng): ")
    if quit_ask in ["yes", "YES", "Yes"]:
        print("Tạm biệt!")
        break
```

```python
# W8A6
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h

    def perimeter(self):
        return (self.w + self.h) * 2

    def scale(self, k):
        self.w *= k
        self.h *= k

w, h, k = map(float, input().split())
rect = Rectangle(w, h)
rect.scale(k)
print(f"{rect.area():.2f} {rect.perimeter():.2f}")
```

```python
# W8A7
import math

class Fraction:
    def __init__(self, num, den):
        if den == 0:
            raise ValueError("Mẫu số phải khác 0")
        common = math.gcd(num, den)
        self.num = num // common
        self.den = den // common
        if self.den < 0:
            self.num = -self.num
            self.den = -self.den

    def add(self, other):
        return Fraction(self.num * other.den + other.num * self.den, self.den * other.den)

    def sub(self, other):
        return Fraction(self.num * other.den - other.num * self.den, self.den * other.den)

    def mul(self, other):
        return Fraction(self.num * other.num, self.den * other.den)

    def div(self, other):
        return Fraction(self.num * other.den, self.den * other.num)

    def __str__(self):
        return f"{self.num}/{self.den}"

parts = input().split()
a, b = int(parts[0]), int(parts[1])
op = parts[2]
c, d = int(parts[3]), int(parts[4])

f1 = Fraction(a, b)
f2 = Fraction(c, d)

if op == '+':
    print(f1.add(f2))
elif op == '-':
    print(f1.sub(f2))
elif op == '*':
    print(f1.mul(f2))
elif op == '/':
    print(f1.div(f2))
```

```python
# W8A8
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = float(balance)

    def deposit(self, x):
        self.balance += float(x)

    def withdraw(self, x):
        x = float(x)
        if x <= self.balance:
            self.balance -= x

    def __str__(self):
        return f"{self.balance:.2f}"

line = input().split()
owner = line[0]
initial_balance = line[1]
acc = BankAccount(owner, initial_balance)
n = int(input())
for _ in range(n):
    cmd_line = input().split()
    action = cmd_line[0]
    amount = cmd_line[1]
    
    if action == "DEPOSIT":
        acc.deposit(amount)
    elif action == "WITHDRAW":
        acc.withdraw(amount)
        
print(acc)
```

```python
# W8A9
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = []

    def add_score(self, subject, score):
        self.scores.append(float(score))

    def avg(self):
        if not self.scores:
            return 0.0
        return sum(self.scores) / len(self.scores)

    def rank(self):
        average = self.avg()
        if average >= 8:
            return "Excellent"
        elif average >= 6.5:
            return "Good"
        elif average >= 5:
            return "Average"
        else:
            return "Poor"

    def display(self):
        print(f"{self.name} {self.avg():.2f} {self.rank()}")

name = input()
n = int(input())

student = Student(name)

for _ in range(n):
    line = input().split()
    if len(line) >= 2:
        subject = line[0]
        score = line[1]
        student.add_score(subject, score)

student.display()
```

```python
# W8A10
class Book:
    def __init__(self, book):
        self.title = book[0]
        self.author = book[1]
        self.year = book[2]
    def __repr__(self):
        return f"{self.title} {self.author} {self.year}"

class Library:
    def __init__(self):
        self.books = []
    def add(self, book):
        self.books.append(book)
    def count_by_author(self, name):
        for book in self.books:
            print(book.author, name)
        return sum([1 for book in self.books if book.author == name])
    def find_by_year(self, y):
        return sum(1 for book in self.books if str(book.year) == str(y))
    def __repr__(self):
        return f"{self.books}"

n = int(input())
Lib = Library()
while n > 0:
    op = input().split()
    op[1] = op[1].split(";")
    # print(op)
    if op[0] == "ADD":
        Lib.add(Book(op[1]))
    elif op[0] == "COUNT":
        print(Lib.count_by_author(op[1][0]))
        # print(op[1], Lib)
    else:
        print(Lib.find_by_year(op[1][0]))
    n -= 1
```

```python
# W8A11
import math

class Vector3D:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
    def __add__(self, other):
        return Vector3D(self.x + other.x, self.y + other.y, self.z + other.z)
    def __sub__(self, other):
        return Vector3D(self.x - other.x, self.y - other.y, self.z - other.z)
    def __mul__(self, other):
        return self.x * other.x + self.y * other.y + self.z * other.z
    def Length(self):
        return math.sqrt(self.x ** 2 + self.y ** 2 + self.z ** 2)
    def Degree(self, other):
        cosin = abs(self * other) / (self.Length() * other.Length())
        radian = math.acos(cosin)
        degree = math.degrees(radian)
        return degree

class Plane:
    def __init__(self, v1, v2):
        self.v1 = v1
        self.v2 = v2
    def VectorOrthogonal(self):
        x1, y1, z1 = self.v1.x, self.v1.y, self.v1.z
        x2, y2, z2 = self.v2.x, self.v2.y, self.v2.z
        x = y1 * z2 - z1 * y2
        y = z1 * x2 - x1 * z2
        z = x1 * y2 - y1 * x2
        if x == 0 and y == 0 and z == 0:
            return "Not a plane"
        return Vector3D(x, y, z)
    def Degree(self, other):
        return self.VectorOrthogonal().Degree(other.VectorOrthogonal())

Points = []
n = 4
while n > 0:
    x, y, z = map(float, input().split())
    Points.append(Vector3D(x, y, z))
    n -= 1

Vectors = []
for i in range(3):
    Vectors.append(Points[i] - Points[i + 1])
Plane1 = Plane(Vectors[0], Vectors[1])
Plane2 = Plane(Vectors[1], Vectors[2])
if type(Plane1.VectorOrthogonal()) == str() or type(Plane2.VectorOrthogonal()) == str():
    print("Error!")
else:
    print(f"{Plane1.Degree(Plane2):.2f}")
```

```python
# W8A12
class ComplexNumber:
    def __init__(self, initRealPart, initImaginaryPart):
        self.realPart = float(initRealPart)
        self.imaginaryPart = float(initImaginaryPart)
    def __mul__(self, other):
        new_real = (self.realPart * other.realPart) - (self.imaginaryPart * other.imaginaryPart)
        new_imag = (self.realPart * other.imaginaryPart) + (self.imaginaryPart * other.realPart)
        return ComplexNumber(new_real, new_imag)

a, b = map(int, input().split())
c1 = ComplexNumber(a, b)

c, d = map(int, input().split())
c2 = ComplexNumber(c, d)
result = c1 * c2

print(f"{result.realPart:g} {result.imaginaryPart:g}")
```

```python
# W8A13
import math

class Triangle:
    def __init__(self, initEdge1, initEdge2, initEdge3):
        self.edge1 = initEdge1
        self.edge2 = initEdge2
        self.edge3 = initEdge3

    def getArea(self):
        a = self.edge1
        b = self.edge2
        c = self.edge3
        if (a + b > c) and (a + c > b) and (b + c > a) and a > 0 and b > 0 and c > 0:
            p = (a + b + c) / 2
            area = math.sqrt(p * (p - a) * (p - b) * (p - c))
            return area
        else:
            return -1

input_data = input().split()
edge1, edge2, edge3 = map(int, input_data)
tri = Triangle(edge1, edge2, edge3)
result = tri.getArea()

if result == -1:
    print("invalid")
else:
    print(f"{result:.2f}")
```
