```python
#W8A3
def sum(n:int):
    s=0
    while(n):
        s+=(n%10)**2
        n//=10
    return s

a=input()
d=set()
kt=0
try:
    x=int(a)
    d.add(x)
    while(x!=1):
        x=sum(x)
        if(x in d):
            kt=1
            print("NO")
            break
        d.add(x)
    if(kt==0):
        print("YES")
except ValueError:
    print("ValueError")
```

```python
#W10A10
a=list(map(int,input().split()))
s=set()
sum=0
for i in range(1,len(a)):
    if(a[i] in s):
        sum=0
        print("Mang khong hop le")
        break
    s.add(a[i])
    sum+=a[i]
if(sum!=0):
    print(sum)
```

```python
#W10A11
def regex(s):
    for i in range(len(s)):
        if s[i] in '*+':
            if i == 0:
                return False
            if s[i-1] in '*+':
                return False
    return True


t = int(input())
for _ in range(t):
    s = input().strip()
    print("True" if regex(s) else "False")
```
