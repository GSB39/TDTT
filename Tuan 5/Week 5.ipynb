```python
# W5A1
def Max_Of_Two( a, b) -> int:
    return max(a, b)
a, b = map(int, input().split())
print(Max_Of_Two(a, b))
```

```python
# W5A2
def swap(a, b):
    return b, a
a, b = map(int, input().split())
print(swap(a, b)[0], swap(a, b)[1])
```

```python
# W5A3
def nt(a):
    if a == 1:
        return False
    for i in range(2, a):
        if a % i == 0:
            return False
    return True
a= int(input())
print(nt(a))
```

```python
# W5A4
def hh(a):
    sum = 0
    for i in range(1, a):
        if a % i == 0:
            sum += i
    return sum == a
a= int(input())
print(hh(a))
```

```python
# W5A5
def check(a, k):
    for i in range(1, len(a)):
        if a[i - 1] == k:
            return i
    return -1
a = list(map(int, input().split()))
k = int(input())
print(check(a, k))
```

```python
# W5A6
def gt(a):
    sum = 1
    for i in range(2, a + 1):
        sum *= i
    return sum
a= int(input())
print(gt(a))
```

```python
# W5A7
import operator
def cal(a):
    op = {"+": operator.add, "-": operator.sub, "*": operator.mul, "/": operator.truediv}
    res = op[a[1]](float(a[0]), float(a[2]))
    return f"{res:.2f}"
a = input().split()
print(cal(a))
```

```python
# W5A8
def Hamming(a, b):
    cnt = 0
    while a > 0 or b > 0:
        cnt += (a & 1) != (b & 1)
        a >>= 1
        b >>= 1
    return cnt
a, b = map(int, input().split())

print(Hamming(a, b))
```

```python
# W5A9
def cal(a):
    return sum(list(map(int, list(a))))
a = input()
print(cal(a))
```

```python
def ax(a, b):
    c = {}
    for i in range(0, len(a)):
        if a[i] in c and c[a[i]] != b[i]:
            # print(a, i + 1)
            return False
        c[a[i]] = b[i]
    return True
a, b = input().split()
print(ax(a, b) and ax(b, a))
```
