```python
#W9A1
def merge(arr1,arr2):
    res=[]
    i,j=0,0
    while(i<len(arr1))and(j<len(arr2)):
        if(arr1[i]>arr2[j]):
            res.append(arr2[j])
            j+=1
        else:
            res.append(arr1[i])
            i+=1
    while(i<len(arr1)):
        res.append(arr1[i])
        i+=1
    while(j<len(arr2)):
        res.append(arr2[j])
        j+=1
    return res
a=list(map(int,input().split()))
b=list(map(int,input().split()))
print(merge(a,b))
```

```python
#W9A4
def kt(s1:str,s2:str):
    i,j=0,0
    while(i<len(s1))and(j<len(s2)):
        if(s1[i]==s2[j])or(ord(s1[i])-32==ord(s2[j])):
            i+=1
            j+=1
        elif("A"<=s1[i]<="Z")and(s1[i]!=s2[j]):
            j=0
            i+=1
        elif(s1[i]!=s2[j])and(ord(s1[i])-32!=ord(s2[j])):
            i+=1
        if(j==len(s2)):
            return "YES"
    return "NO"

q=int(input())
while(q):
    s1=input()
    s2=input()
    print(kt(s1,s2))
    q-=1
```

```python
#W9A9
from bisect import bisect_left
def solve(houses, heaters):
    houses.sort()
    heaters.sort()
    res = 0
    for h in houses:
        idx = bisect_left(heaters, h)
        dist_left = float('inf')
        dist_right = float('inf')
        if idx > 0:
            dist_left = h - heaters[idx - 1]
        if idx < len(heaters):
            dist_right = heaters[idx] - h
        nearest = min(dist_left, dist_right)
        res = max(res, nearest)
    return res
house = list(map(int, input().split()))
heater = list(map(int, input().split()))
print(solve(house, heater))
```
