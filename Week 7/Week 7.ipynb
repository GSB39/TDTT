```python
# W7A1
a = list(map(int, input().split()))
k = int(input())
l, r = 0, len(a) - 1
while l < r:
    mid = (l + r) // 2
    if a[mid] < k:
        l = mid + 1
    else:
        r = mid
if a[l] == k:
    print(l)
else:
    print(-1)
```

```python
# W7A2
def count_occurrences(arr, x) -> int:
    cnt = {}
    for i in arr:
        if i not in cnt:
            cnt[i] = 1
        else:
            cnt[i] += 1
    return cnt[x] if x in cnt else 0
a = list(map(int, input().split()))
x = int(input())
print(count_occurrences(a, x))
```

```python
# W7A3
def selection_sort(a):
    n = len(a)
    for i in range(n):
        for j in range(i + 1, n):
            if a[i] > a[j]:
                a[i], a[j] = a[j], a[i]
def bubble_sort(a):
    n = len(a)
    for i in range(n - 1, -1, -1):
        for j in range(0, i):
            if a[j] > a[j + 1]:
                a[j], a[j + 1] = a[j + 1], a[j]
a = list(map(int, input().split()))
bubble_sort(a)
for x in a:
    print(x, end = " ")
```

```python
# W7A4
def max_exist(a):
    cnt = {}
    for i in range(len(a)):
        if a[i] not in cnt:
            cnt[a[i]] = (1, i)
        else:
            cnt[a[i]] = (cnt[a[i]][0] + 1, i)
    res = (0, 0, 0)
    for val, (ex, pos) in cnt.items():
        res = max(res, (ex, -pos, val))
    return res
a = list(map(int, input().split()))
ans = max_exist(a)
print(f"{ans[2]} xuat hien nhieu nhat, som nhat, {ans[0]} lan")
```

```python
# W7A7
a = list(map(int, input().strip("[]").split(",")))
n = len(a)
tang = [min(a) - 1] + [max(a) + 1] * n
res = 0
for i in range(n):
    l, r = 0, i
    while l < r:
        mid = l + r + 1 >> 1
        if tang[mid] >= a[i]:
            r = mid - 1
        else:
            l = mid
    if tang[l + 1] > a[i]:
        tang[l + 1] = a[i]
        res = max(res, l + 1)
print(res)
```

```python
# W7A7
import ast
import heapq
def get_min(h):
    if not h:
        return -1
    return h[0]
def add(h, val):
    heapq.heappush(h, val)
def remove(h):
    if not h:
        return None
    heapq.heappop(h)
a = ast.literal_eval(input())
q = ast.literal_eval(input())
c = []
val = []
cnt = {}
for [l, r] in a:
    c.append((l, 0, r - l + 1))
    c.append((r, 2, r - l + 1))
    cnt[r - l + 1] = 0
for i in range(len(q)):
    c.append((q[i], 1, i))
c.sort()

res = [-1 for i in range(len(q))]
for (pos, t, leng) in c:
    print(pos, leng, -1)
    if t == 0:
        add(val, leng)
        cnt[leng] += 1
    elif t == 2:
        cnt[leng] -= 1
        while len(val) != 0 and cnt[get_min(val)] == 0:
            remove(val)
    else:
        res[leng] = get_min(val)
        print(pos, leng, res[leng])
print(res)
```

```python
# W7A9
a = list(map(int, input().strip("[]").split(",")))
m = max(a) + 2 # Code không có giới hạn nên lấy tạm giá trị
BIT = [0] * m
def get(x):
    sum, i = 0, x
    while i > 0:
        sum += BIT[i]
        i -= i & -i
    return sum
def update(x, val):
    i = x
    while i <= m:
        BIT[i] += val
        i += i & -i
res = []
for i in range(len(a) - 1, -1, -1):
    res.append(get(a[i]))
    update(a[i], 1)
res.reverse()
print(res)
```

```python
# W7A10
a = list(map(int, input().strip("[]").split(",")))
mod = int(input())
a = [0] + a
pre = 0
mi, mx = 0, 0
m = mod + 3 # Code không có giới hạn nên lấy tạm giá trị
BIT = [0] * m
def get(x):
    sum, i = 0, x
    while i <= m:
        sum += BIT[i]
        i += i & -i
    return sum
def update(x, val):
    i = x
    while i > 0:
        BIT[i] += val
        i -= i & -i
res = 0
for i in range(1, len(a)):
    pre += a[i]
    pre %= mod
    update(pre, 1)
    mi = min(mi, pre)
    mx = pre
    l, r = pre + 1, mod - 1
    maxex = get(pre + 1)
    while l < r:
        mid = l + r + 1 >> 1
        if get(mid) == maxex:
            l = mid
        else:
            r = mid - 1
    if get(l) == maxex and maxex != 0:
        mx = l
    # print(i, pre, l, maxex)
    res = max(res, pre - mi, (pre - mx) % mod)
    
print(res)
```

```python
# W7A11
t = input()
s = input()
def prefix_function(s):
    n = len(s)
    pi = [0] * n
    for i in range(1, n):
        j = pi[i - 1]
        while j > 0 and s[i] != s[j]:
            j = pi[j - 1]
        if s[i] == s[j]:
            j += 1
        pi[i] = j
    return pi

pi = prefix_function(s)
ans = 0
j = 0
for i in range(len(t)):
    while j > 0 and t[i] != s[j]:
        j = pi[j - 1]
    if t[i] == s[j]:
        j += 1
    if j == len(s):
        ans += 1
        j = pi[j - 1]
print(ans)
```
