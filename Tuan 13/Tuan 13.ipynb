```python
#W11A1
def reverse(path: str) -> int:
    with open(path, 'r') as f:
        n = f.read().strip()
    return int(n[::-1])

path = input().strip()
print(reverse(path))
```

```python
#W11A2
def canWinNim(path: str) -> bool:
    with open(path, 'r') as f:
        n = int(f.read().strip())
    return n % 4 != 0

path = input().strip()
result = canWinNim(path)
print("true" if result else "false")
```

```python
#W11A4
def normal(s):
    return sorted(c for c in s.lower() if c.isalpha())
def areAnagrams(path1: str, path2: str):
    with open(path1, 'r') as f:
        s1 = f.read()
    with open(path2, 'r') as f:
        s2 = f.read()
    return normal(s1) == normal(s2)

path1 = input().strip()
path2 = input().strip()
print("True" if areAnagrams(path1, path2) else "False")
```

```python
#W11A5
def count(path: str, character: str) -> int:
    with open(path, 'r') as f:
        sentence = f.read()
    return sentence.count(character)

path = input().strip()
character = input().strip()
print(count(path, character))
```

```python
#W11A6
def countCharacter(path: str, letter: str) -> int:
    with open(path, 'r') as f:
        text = f.read()
    return text.lower().count(letter.lower())

path = input().strip()
letter = input().strip()
print(countCharacter(path, letter))
```

```python
#W11A10
def getMoney(path: str):
    with open(path, 'r') as file:
        data = file.read().strip()
    if data == "":
        return []
    return list(map(int, data.split()))

def rob(hourses):
    if not hourses:
        return 0
    n = len(hourses)
    if n == 1:
        return hourses[0]
    if n == 2:
        return max(hourses[0], hourses[1])
    dp = [0] * n
    dp[0] = hourses[0]
    dp[1] = max(hourses[0], hourses[1])
    for i in range(2, n):
        dp[i] = max(dp[i - 1], dp[i - 2] + hourses[i])
    return dp[-1]

path = input()
houses = getMoney(path)
result = rob(houses)
print(result)
```

```python
#W11A12
def maxProduct(path:str) -> int:
    with open(path, 'r') as file:
        n = int(file.readline().strip())
        arr = list(map(int, file.readline().split()))
    arr.sort()
    product1 = arr[-1] * arr[-2] * arr[-3]
    product2 = arr[0] * arr[1] * arr[-1]
    return max(product1, product2)
path = input()
print(maxProduct(path))
```

```python
#W11A13
def moveZeros(fileName: str) -> list[int] :
    with open(fileName, "r") as f :
        n = int(f.readline().strip())
        lst = list(map(int,f.readline().split()))
    cnt = 0
    result = []
    for i in lst :
        if i != 0 :
            result.append(i)
        else :
            cnt += 1
    result.extend([0] * cnt)
    return result
fileName = input()
print(moveZeros(fileName))
```

```python
#W11A14
def averageTime(path: str):
    total = 0.0
    count = 0
    with open(path, 'r') as file:
        next(file)
        for line in file:
            lap, time = line.strip().split(',')
            total += float(time)
            count += 1
    return round(total / count, 2)

path = input()
result = averageTime(path)
print(result)
```

```python
#W11A15
def findMovies(path: str) -> list[int]:
    result = []
    with open(path, 'r') as file:
        next(file)
        for line in file:
            parts = [x.strip() for x in line.split(',')]
            movie = parts[1]
            description = parts[2]
            rating = float(parts[3])
            if rating > 8.0 and description != "boring":
                result.append(movie)
    if not result:
        return "All are boring!"
    return result

path = input()
output = findMovies(path)
if isinstance(output, list):
    for movie in output:
        print(movie)
else:
    print(output)
```

```python
#W11A16
import math
path = input()
with open(path, 'r') as file:
    lines = [line.strip() for line in file if line.strip() != ""]
i = 0
while i < len(lines):
    shape_type = lines[i].split(":")[1].strip()
    params = lines[i + 1]
    if shape_type == "SQUARE":
        side = float(params.split(":")[1])
        perimeter = 4 * side
    elif shape_type == "RECTANGLE":
        parts = params.split()
        width = float(parts[1])
        height = float(parts[3])
        perimeter = 2 * (width + height)
    elif shape_type == "CIRCLE":
        radius = float(params.split(":")[1])
        perimeter = 2 * math.pi * radius
    print(f"{perimeter:.2f}")
    i += 2
```

```python
#W11A17
def productExceptSelf(path: str) -> list[int]:
    with open(path, 'r') as file:
        arr = list(map(int, file.read().split()))
    n = len(arr)
    left = [1] * n
    right = [1] * n
    result = [1] * n
    for i in range(1, n):
        left[i] = left[i - 1] * arr[i - 1]
    for i in range(n - 2, -1, -1):
        right[i] = right[i + 1] * arr[i + 1]
    for i in range(n):
        result[i] = left[i] * right[i]
    return result
path = input()
res = productExceptSelf(path)
print(*res)
```

```python
#W11A18
import math
def estimatedTime(path: str) -> float:
    with open(path, "r") as f:
        lines = f.readlines()
    src_parts = lines[0].split()
    x1 = float(src_parts[2])
    y1 = float(src_parts[4])
    des_parts = lines[1].split()
    x2 = float(des_parts[2])
    y2 = float(des_parts[4])
    velocity = float(lines[2].split()[1])
    distance = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
    time = distance / velocity
    return round(time, 2)
path = input()
print(estimatedTime(path))
```

```python
#W11A20
def imageSmoother(path: str):
    with open(path, "r") as f:
        M = int(f.readline().strip())
        img = []
        for _ in range(M):
            img.append(list(map(int, f.readline().split())))
    result = [[0] * M for _ in range(M)]
    directions = [(-1, -1),(-1, 0),(-1, 1),(0, -1),(0, 0), (0, 1),(1, -1),(1, 0),(1, 1)]
    for i in range(M):
        for j in range(M):
            total = 0
            count = 0
            for dx, dy in directions:
                ni, nj = i + dx, j + dy
                if 0 <= ni < M and 0 <= nj < M:
                    total += img[ni][nj]
                    count += 1
            result[i][j] = total // count
    for row in result:
        print(*row)
```
